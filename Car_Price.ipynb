# One Hot Encoding

```python
droped_coloumns=['car_ID','symboling','CarName','fuelsystem','carlength','wheelbase']
df=Car_df.drop(columns=droped_coloumns,axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder

X = (df.dtypes == 'object')
Obj_Num = list(X[X].index)
print("Objects : ")
print(Obj_Num)
print('Number of Features : ',
   len(Obj_Num))
```

```python
Encoder = OneHotEncoder(sparse_output=False)
Cols = pd.DataFrame(Encoder.fit_transform(df[Obj_Num]))
Cols.index = df.index
Cols.columns = Encoder.get_feature_names_out()
Output_df = df.drop(Obj_Num, axis=1)
Output_df = pd.concat([Output_df, Cols], axis=1)
```

```python
Output_df.shape
```

```python
Output_df.head(10)
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()


scaled_Output = scaler.fit_transform(Output_df)
scaled_Output
```

```python
Result_df=pd.DataFrame(scaled_Output)
Result_df.columns=Output_df.columns
Result_df.head(10)
```

# Testing Traning & Spliting

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
 
X = Result_df.drop(['price'], axis=1)
Y = Result_df['price']
 

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, train_size=0.7, test_size=0.3, random_state=42)
```

```python
X_train
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
 
Linear_Model = LinearRegression()
Linear_Model.fit(X_train, Y_train)
Y_pred = Linear_Model.predict(X_test)
 
```

```python
mse=mean_squared_error(Y_test, Y_pred)
mse
```

```python
import numpy as np
Y=np.sqrt(mean_squared_error(Y_test, Y_pred))
Y
```
